```
fdist1 = FreqDist(text1)
fdist1.most_common(10)
```

```
# Checking zipf's law

import matplotlib.pyplot as plt
def zipf(l):
  L = []
  for i in fdist1.most_common(l):
    L.append(i[1])
  return L  
plt.plot(zipf(100))
```

```
def vocab_size(text):
  S = {w.lower() for w in text}
  return len(S) 

vocab_size(text1)
```

```
def percent(word, text):
  i = 0
  for w in text:
    if w == word: i += 1
  return round(i/len(text)*100,2)

percent('the',text3)
```
